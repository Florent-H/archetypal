```python
thermal_zones = pd.DataFrame.from_dict(data["thermal_zones"])
spaces = pd.DataFrame.from_dict(data["spaces"])
```

```python
def ach(row):
    """
    Air changes per hour
    """
    q = row['infiltration_flow_m3_per_s']
    V = row['volume']
    if q > 0:
        return 3600 * q / V
    return np.NaN
```

```python
spaces['infiltration_flow_m3_per_s'] = spaces['infiltration_flow_per_m2'] * spaces['exterior_wall_area']
```

```python
spaces['air_changes_per_hour'] = spaces.apply(lambda x: ach(x), axis=1)
```

```python
spaces
```

```python
# Define a lambda function to compute the weighted mean:
wm = lambda x: np.average(x, weights=(spaces.loc[x.index, 'volume'] * spaces.loc[x.index, 'multiplier']))
wm_2 = lambda x: np.average(x, weights=(spaces.loc[x.index, 'volume'] * spaces.loc[x.index, 'multiplier']))

# Define a dictionary with the functions to apply for a given column:
f = {'air_changes_per_hour': {'weighted_mean' : wm_2}, 'occ_per_m2' : {'weighted_mean' : wm} }

# Groupby and aggregate with your dictionary:
spaces.agg(f)
```

```python
# Reading the json as a dict
datadict = {}
for file in files:
    with open(file) as json_data:
        data = json.load(json_data)
    datadict[file] = data["end_uses_eui"]
```

```python
energytable = pd.DataFrame.from_dict(datadict).filter(like='gj', axis=0).applymap(lambda x: x/0.0036)
```

```python
energytable
```

```python
energytable.T.to_excel('energy-table.xlsx')
```

```python
## Equip : Equip_ext + Equip_int + Fan + Pump
eui_table = energytable.T

# Equip : Equip_ext + Equip_int + Fan + Pump
eui_table['equipement_kwh_per_m2'] = eui_table['exterior_equipment_gj_per_m2'] + eui_table['interior_equipment_gj_per_m2'] + eui_table['fans_gj_per_m2'] + eui_table['pumps_gj_per_m2']
eui_table.drop(labels=['exterior_equipment_gj_per_m2', 'interior_equipment_gj_per_m2', 'fans_gj_per_m2', 'pumps_gj_per_m2'], axis=1, inplace=True)

# Heating : 
eui_table['heating_kwh_per_m2']
```
